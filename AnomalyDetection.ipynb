```python
from pycaret.utils import enable_colab
enable_colab()
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Chaitra-B-V/AnamolyDetectionUsingPycaret/main/ChillerData.csv'

dataset = pd.read_csv(url,nrows=1500)
```

```python
#check the shape of data
dataset.shape
```

```python
data = dataset.sample(frac=0.95, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.anomaly import *

exp_ano101 = setup(data, normalize = True, use_gpu=True,ignore_features = ['Date'],
                   session_id = 123)
```

```python
iforest = create_model('iforest')
```

```python
print(iforest)
```

```python
svm = create_model('svm', fraction = 0.025)
```

```python
print(svm)
```

```python
models()
```

```python
iforest_results = assign_model(iforest)
iforest_results.head()
```

```python
plot_model(iforest)
```

```python
plot_model(iforest, plot = 'umap')
```

```python
unseen_predictions = predict_model(iforest, data=data_unseen)
unseen_predictions.head()
```

```python
data_predictions = predict_model(iforest, data = data)
data_predictions.head()
```

```python
save_model(iforest,'Final IForest AD')
```

```python
saved_iforest = load_model('Final IForest AD')
```

```python
new_prediction = predict_model(saved_iforest, data=data_unseen)
```
